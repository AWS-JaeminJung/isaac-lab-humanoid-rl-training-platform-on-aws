# Submit Training via OSMO Client

This notebook demonstrates how to submit Isaac Lab training jobs through the
OSMO orchestrator using the `osmo-client` Python SDK.

## Prerequisites
- OSMO API endpoint accessible (set via `OSMO_API_URL` env var)
- Valid Keycloak authentication token
- Training configuration YAML prepared

```python
import os
from osmo_client import OsmoClient

# Initialize the OSMO client
client = OsmoClient(
    api_url=os.environ.get("OSMO_API_URL", "http://osmo-controller.osmo.svc.cluster.local:8080"),
)
print(f"Connected to OSMO API: {client.api_url}")
```

## Define Training Configuration

Specify the Isaac Lab environment, algorithm, and resource requirements.

```python
# Define training job configuration
training_config = {
    "name": "cartpole-ppo-experiment-001",
    "environment": "Isaac-Cartpole-v0",
    "algorithm": "PPO",
    "num_envs": 4096,
    "max_iterations": 500,
    "resources": {
        "gpu": 1,
        "cpu": 4,
        "memory": "16Gi",
    },
    "hyperparameters": {
        "learning_rate": 3e-4,
        "gamma": 0.99,
        "batch_size": 2048,
        "num_epochs": 5,
    },
}

print("Training configuration:")
for key, value in training_config.items():
    print(f"  {key}: {value}")
```

```python
# Submit the training job
job = client.submit_training(training_config)

print(f"Job submitted successfully!")
print(f"  Job ID:   {job.id}")
print(f"  Status:   {job.status}")
print(f"  Created:  {job.created_at}")
```

```python
# Check job status
status = client.get_job_status(job.id)
print(f"Job {job.id}: {status.state}")
print(f"  Progress: {status.progress}%")
print(f"  Duration: {status.duration}s")
```
